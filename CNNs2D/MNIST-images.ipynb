#### Load MNIST dataset ####
* Torchvision provides many datasets including MNIST

```python
import os
from torchvision.datasets import MNIST
from torchvision import transforms
import numpy as np
import matplotlib.pyplot as plt

transform = transforms.Compose([transforms.ToTensor(), transforms.Normalize((0.1307), (0.3081))])
mnist = MNIST(os.getcwd(), train=True,  download=True, transform=transform)

print(mnist)
```

#### MNIST data ####
* An example
* 1 sample is composed of a label and an image(28 pixel x 28 pixes)

```python
image, label = mnist[0]
print('label:', label)
print('size of image:', image.shape)

plt.imshow(image.numpy()[0], cmap='gray')
```

#### Check 40 samples ####

```python
figs, axes = plt.subplots(4,10,figsize=[32,12])

for i in range(40):
    r = int(i/10)
    c = int(i%10)
    image, label = mnist[i]
    axes[r,c].set_title('{}'.format(label), weight='bold')
    axes[r,c].imshow(image.numpy()[0], cmap='gray')
```
